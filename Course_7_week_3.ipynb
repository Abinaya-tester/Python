```python
import pandas as pd

data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'City': ['New York', 'London', 'New York', 'London'],
    'Temperature': [32, 45, 30, 42]
}
df = pd.DataFrame(data)
```

```python
pivoted = df.pivot(index='Date', columns='City')
print(pivoted)
```

```python
import pandas as pd

data1 = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'City': ['New York', 'London', 'New York', 'London'],
    'Temperature': [33, 45, 30, 42],
    'Humidity':[10,11,12,13]
}
df1 = pd.DataFrame(data1)

pivoted = df1.pivot(index='Date', columns='City', values='Temperature')
print(pivoted)
```
